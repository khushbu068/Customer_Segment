Create new features for clustering and Select relevant features for clustering.
Summing purchases across all brands

```python
data['Total Orders'] = data['Orders']
data['Total Brands Purchased'] = data.iloc[:, 3:].sum(axis=1)
features = data[['Total Orders', 'Total Brands Purchased']]
print("Selected features for clustering:\n", features.head())
```

Determine the optimal number of clusters using the elbow method

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

Apply K-means with the chosen number of clusters and Save the clustered data

```python
optimal_clusters = 3  
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
data['Cluster'] = kmeans.fit_predict(scaled_features)
clustered_file_path = r'C:\Users\madha\Desktop\Customer_segment_project\clustered_customer_data_.csv'
data.to_csv(clustered_file_path, index=False)
```

Silhouette Score
Pair plot for clusters using the correct column names and printing.

```python
score = silhouette_score(scaled_features, data['Cluster'])
print(f'Silhouette Score: {score}')
 
sns.pairplot(data, hue='Cluster', vars=['Total Orders', 'Total Brands Purchased'])
plt.show()
print("Cluster centers:\n", kmeans.cluster_centers_)
```
